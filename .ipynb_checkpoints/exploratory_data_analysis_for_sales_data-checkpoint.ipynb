## Random forest

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
from sklearn.model_selection import cross_val_score
from sales_data_extractor import DatabaseConnector


class SalesAnalyzer:
    def __init__(self, config_file):
        self.db_connector = DatabaseConnector(config_file)
        self.x_columns = []  # Initialize x_columns
        self.y_column = ""  # Initialize y_column

    def fetch_sales_data(self, query):
        query_results = self.db_connector.execute_query(query)
        results, column_names = query_results
        self.df = pd.DataFrame(results, columns=column_names)

    def select_best_predictors(self, x_columns, y_column, n_features_to_select=2):
        self.x_columns = x_columns  # Store x_columns
        self.y_column = y_column  # Store y_column

        x = self.df[x_columns]
        y = self.df[y_column]

        model = RandomForestRegressor(n_estimators=100)

        rfe = RFE(model, n_features_to_select=n_features_to_select)
        rfe = rfe.fit(x, y)

        top_predictors = [col for col, support in zip(x_columns, rfe.support_) if support]

        return top_predictors

    def clean_data(self):
        self.df[self.x_columns] = self.df[self.x_columns].apply(pd.to_numeric, errors='coerce')
        self.df[self.y_column] = pd.to_numeric(self.df[self.y_column], errors='coerce')

    def perform_regression_analysis(self):
        model = RandomForestRegressor(n_estimators=100)
        x = self.df[self.x_columns]
        y = self.df[self.y_column]

        # Use cross-validation to assess model performance
        scores = cross_val_score(model, x, y, cv=5, scoring='neg_mean_squared_error')

        # Calculate the RMSE based on the cross-validation scores
        rmse = np.sqrt(-scores.mean())  # Take the negative mean squared error
        print(f"Cross-Validated RMSE: {rmse:.2f}")

        # Fit the model on the entire dataset
        model.fit(x, y)
        self.model = model

    def evaluate_model(self):
        actual_values = self.df[self.y_column]
        predicted_values = self.model.predict(self.df[self.x_columns])

        rmse = np.sqrt(mean_squared_error(actual_values, predicted_values))
        mae = mean_absolute_error(actual_values, predicted_values)
        mape = mean_absolute_percentage_error(actual_values, predicted_values)
        r_squared = r2_score(actual_values, predicted_values)

        print("Predicted vs. Actual Values:")
        for i in range(0, 5):
            print(f"expected={actual_values[i]:.1f}, predicted={predicted_values[i]:.1f}")

        print(f"RMSE: {rmse:.2f}")
        print(f"MAE: {mae:.2f}")
        print(f"MAPE: {mape:.2f}%")
        print(f"R-squared: {r_squared:.2f}")

    def visualize_predictions(self, num_data_points):
        self.df['ORDERDATE'] = pd.to_datetime(self.df['ORDERDATE'])  # Convert ORDERDATE to datetime
        actual_values = self.df[self.y_column]
        predicted_values = self.model.predict(self.df[self.x_columns])

        # Filter the data to show only the first 'num_data_points' values
        actual_values = actual_values[:num_data_points]
        predicted_values = predicted_values[:num_data_points]

        plt.figure(figsize=(12, 6))
        plt.plot(self.df['ORDERDATE'][:num_data_points], actual_values, label='Actual', marker='o', markersize=4,
                 linewidth=0.5)
        plt.plot(self.df['ORDERDATE'][:num_data_points], predicted_values, label='Predicted', marker='o', markersize=4,
                 linewidth=0.5)

        plt.title = ("Actual vs. Predicted Sales Over Time")
        plt.xlabel = ("Date")
        plt.ylabel = ("Sales")
        plt.legend()
        plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
        plt.show()

    def close_connection(self):
        self.db_connector.close_connection()


if __name__ == "__main__":
    user_query = "SELECT * FROM sales"
    config_file = 'config.json'  # Specify the config file here
    sales_analyzer = SalesAnalyzer(config_file)
    sales_analyzer.fetch_sales_data(user_query)

    x_columns = ['QUANTITYORDERED', 'PRICEEACH']
    y_column = 'SALES'

    top_predictors = sales_analyzer.select_best_predictors(x_columns, y_column, n_features_to_select=2)

    sales_analyzer.clean_data()
    sales_analyzer.perform_regression_analysis()
    sales_analyzer.evaluate_model()
    sales_analyzer.visualize_predictions(num_data_points=25)

    sales_analyzer.close_connection()
````

## Initializing the SalesAnalyzer Class

We define a Python class, `SalesAnalyzer`, which will encapsulate the entire sales analysis process. The class is initialized with a configuration file for database connection and initializes variables for data columns.

## Fetching Sales Data

The `fetch_sales_data` method retrieves sales data from a database based on a user-provided query. The data is stored in a Pandas DataFrame for further analysis.

## Selecting Best Predictors

The `select_best_predictors` method performs feature selection to identify the most important predictors for the sales prediction. It uses a Random Forest Regressor with Recursive Feature Elimination (RFE) to make this selection.

## Cleaning Data

Data cleaning is essential for any analysis. The `clean_data` method converts relevant columns to numeric data types and handles any data inconsistencies.

## Performing Regression Analysis

Originally, the script attempted to use Linear Regression for sales prediction. However, it was found that Linear Regression assumptions were not met, leading to less accurate predictions. Consequently, the script transitions to using a Random Forest Regressor, a non-linear machine learning algorithm known for its robustness and ability to handle complex relationships in the data.

The `perform_regression_analysis` method applies the Random Forest Regressor to predict sales values. It uses cross-validation to assess model performance and calculates the Root Mean Squared Error (RMSE) based on cross-validation results.

## Evaluating the Model

The `evaluate_model` method assesses the model's predictive accuracy. It calculates various metrics such as RMSE, Mean Absolute Error (MAE), Mean Absolute Percentage Error (MAPE), and R-squared (R). It also displays a sample of predicted vs. actual values.

## Visualizing Predictions

To visualize the sales predictions over time, the `visualize_predictions` method creates a time series plot. It allows users to specify the number of data points to display in the graph.

## Closing the Database Connection

The `close_connection` method ensures a clean exit by closing the database connection.

## Running the Analysis

In the main part of the script, a user query is defined to select sales data from the database. The `SalesAnalyzer` class is instantiated with a configuration file, and the analysis is executed step by step:

1. Sales data is fetched.
2. Data is cleaned.
3. Best predictors are selected.
4. A Random Forest regression analysis is performed (switched from Linear Regression due to failed assumptions).
5. The model is evaluated.
6. Predictions are visualized for the first 25 data points.


# Evaluating Model Performance

When assessing the performance of a predictive model, several metrics are commonly used to determine its effectiveness in capturing and predicting the underlying patterns in the data. In this analysis, we have considered several key metrics:

1. **R-squared (R)**: R-squared is a statistical measure of the proportion of the variance in the dependent variable that is explained by the independent variables in a regression model. A high R value, close to 1, indicates that a large percentage of the variance in the data has been accounted for by the model. In this case, we have an R value of 0.83, which is relatively high. This suggests that approximately 83% of the variance in sales data can be explained by our predictive model, indicating that the model captures a significant portion of the underlying relationships.

2. **Root Mean Squared Error (RMSE)**: RMSE is a measure of the average magnitude of the errors between predicted and actual values. A lower RMSE indicates that the model's predictions are closer to the actual values, which is desirable. Here, we have an RMSE of 753.42, which is relatively low. This suggests that, on average, our model's predictions are close to the actual sales values, signifying its accuracy.

3. **Mean Absolute Percentage Error (MAPE)**: MAPE measures the percentage difference between predicted and actual values. A low MAPE implies that the model's predictions are highly accurate. In this analysis, we have an MAPE of 0.08%, indicating that the model's predictions are, on average, very close to the actual sales figures.

4. **Mean Absolute Error (MAE)**: MAE is a measure of the average magnitude of errors between predicted and actual values, similar to RMSE. A low MAE suggests that the model's predictions are close to the actual values. Here, we have an MAE of 395.10, which is relatively low. It means that, on average, the model's predictions deviate from the actual values by a small amount.

### Why the Model Is Considered Substantial

The high R-squared value (0.83) indicates that our model explains a substantial portion of the variance in sales data. Additionally, the low RMSE, MAPE, and MAE values (753.42, 0.08%, and 395.10, respectively) demonstrate that the model's predictions closely align with the actual sales figures. 

In summary, the high R-squared value suggests that the model captures a significant amount of the underlying patterns, while the low RMSE, MAPE, and MAE values reflect the model's ability to make accurate predictions. These metrics collectively indicate that the model is good at explaining and predicting sales data, making it a reliable tool for decision-making and forecasting.
